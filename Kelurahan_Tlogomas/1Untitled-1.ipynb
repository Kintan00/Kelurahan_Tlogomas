```python
import pandas as pd

# 1. Baca file Excel
file_path = r"C:\Users\DELL\Documents\Kelurahan_Tlogomas\Data_Usaha_Tlogomas.xlsx"
df = pd.read_excel(file_path)

# 2. Cek apakah kolom Latitude dan Longitude ada
if 'Latitude' in df.columns and 'Longitude' in df.columns:
    # 3. Buat kolom Coordinates hanya jika nilai Latitude dan Longitude bukan "-"
    df['Coordinates'] = df.apply(
        lambda row: f"{row['Latitude']}, {row['Longitude']}" 
        if str(row['Latitude']).strip() != "-" and str(row['Longitude']).strip() != "-" 
        else "", axis=1
    )

    # 4. Simpan ke file baru (optional)
    output_path = r"C:\Users\DELL\Documents\Kelurahan_Tlogomas\Data_Usaha_Coordinates.xlsx"
    df.to_excel(output_path, index=False)

    print(f"✅ Sukses! File baru disimpan di:\n{output_path}")
else:
    print("❌ Kolom 'Latitude' dan 'Longitude' tidak ditemukan di file.")

df
```
