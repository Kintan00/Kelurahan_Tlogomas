```python
import pandas as pd
import re

# Load file
df = pd.read_excel("C:/Users/DELL/Documents/Kelurahan_Tlogomas/Data_Usaha_Tlogomas.xlsx")

# Ambil kolom kategori
kategori_asli = df['Kategori Lapangan Usaha'].dropna().unique()

# Buat mapping awalan lama ke huruf baru (A, B, C, ...)
huruf_abjad = [chr(i) for i in range(65, 65 + len(kategori_asli))]  # ['A', 'B', 'C', ...]
kategori_map = dict(zip(kategori_asli, huruf_abjad))

# Fungsi untuk mengganti awalan kode dengan huruf
def ganti_awalan(teks):
    for k, v in kategori_map.items():
        if teks.strip() == k:
            # Ambil bagian setelah titik pertama (tanpa huruf awal lama)
            bagian_setelah_titik = re.sub(r'^[A-Z]\.\s*', '', teks.strip())
            return f"{v}. {bagian_setelah_titik}"
    return teks

# Tambahkan kolom baru berisi kode huruf kategori
df['Kode Kategori'] = df['Kategori Lapangan Usaha'].map(kategori_map)

# Terapkan ke kolom
df['Kategori Lapangan Usaha'] = df['Kategori Lapangan Usaha'].apply(ganti_awalan)

# Simpan ke file baru
df.to_excel("C:/Users/DELL/Documents/Kelurahan_Tlogomas/Data_Usaha_Tlogomas.xlsx", index=False)
# Tampilkan mapping
print("Mapping Awalan Baru:")
for asli, huruf in kategori_map.items():
    print(f"{huruf}. {asli}")

```

```python
import pandas as pd
import re

# Baca data dari Excel
df = pd.read_excel("Data_Usaha_Tlogomas.xlsx")

# Gabungkan Nama Usaha + Deskripsi jika ada
df['Gabungan'] = df[['Nama Usaha', 'Deskripsi Aktifitas']].fillna('').agg(' '.join, axis=1).str.lower()

# Fungsi klasifikasi berdasarkan kata kunci
def klasifikasi_jenis_usaha(text):
    # Kos
    if re.search(r'\bkos\b|\bkost\b|\bkostan\b', text):
        return "Kost"
    if re.search(r'\bwisma\b|\bkontrakan\b', text):
        return "Penginapan  / Kontrakan"
    # Toko Kelontong / Sembako / Minimarket
    elif re.search(r'\btoko\b.*(kelontong|lampu|sembalo|perlengkapan|klontong|sayur|elpigi|helm|mart|seragam|darda|botol|plastik|sembako|serba|serbi|kebutuhan)|menjual.*sembako', text):
        return "Toko Kelontong / Sembako/ Minimarket"
    # Warung Makan / Makanan Berat
    elif re.search(r'\bwarung\b|\bkedai\b|\bgeprek\b|\bayam\b|\bnasi\b|\bmi\b|\bpadang\b|\bakso\b|\bkatering\b|\bcatering\b|\bbolu\b|\bmie\b|\bakso\b|\bwarmindo\b|\bsoto\b|\bsambel\b|\blauk\b|\bmasakan\b|\bpecel\b', text):
        return "Warung Makan / Makanan Berat"
    elif re.search(r'\bhp\b|\bhandphone\b|\bpulsa\b|\blaptop\b|\baksesori\b|\bgadget\b|\bcell\b|\bmac\b|\bcharger\b|\bperangkat\b|\bteknologi\b|\bservice\b|\bservis\b|\bcomputer\b', text):
        return  "Elektronik / Servis Gadget"
    elif re.search(r'\bhewan\b|\baquatic\b|\bpakan\b|\baquana\b|\bakuarium\b|\bpet\b|\banjing\b|\bkucing\b|\bfish\b|\baquatic\b', text):
        return "Pet Shop / Akuarium"
    # Jajanan / Camilan / Martabak
    elif re.search(r'jajanan|pempek|roti|rengginang|crepes|seblak|simbok|gorengan| kue | tahu |kripik|mendoan|cakwe |martabak|cilok|pisang|keripik|kue|\bbakery\b|takoyaki|cimol|basreng|telur|serabi|rujak|snack', text):
        return "Penjual Jajanan / Camilan"
    # Minuman / Es / Juice / Teh
    elif re.search(r'\bminuman\b|\bes\b|juice|jus|stmj|jamu|teh|teler', text):
        return "Penjual Minuman"
    # Salon / Kecantikan / MUA
    elif re.search(r'salon|rias|rambut|hair cut|kecantikan|makeup|mua|beauty|spa|perawatan', text):
        return "Salon / Kecantikan"
    elif re.search(r'\bsalon\b|\bbarber\b|\bbarbershop\b|\bmakeup\b|\bspa\b|\bnail\b|\bkuku\b|\bhair\b|\bmoist\b|\btreatment\b|\bkecantikan\b|\bkosmetik\b|\bmua\b', text):
        return "Salon / Kecantikan"
    # Laundry / Cuci Pakaian / Self Service
    elif re.search(r'laundry|cuci pakaian|cuci|clean|laundry self', text):
        return "Laundry"
    # Fotocopy / Printing / Percetakan
    elif re.search(r'fotocopy|printing|design|percetakan|print|jilid', text):
        return "Fotocopy / Printing / Percetakan"
    # Penjahit / Permak / Konveksi
    elif re.search(r'jahit|vermak|penjahit|konveksi', text):
        return "Penjahit / Permak / Konveksi"
    # Bengkel / Service Motor / Mobil
    elif re.search(r'bengkel|servis motor|servis mesin|cuci motor|servis mobil|reparasi|ahass|knalpot|ban|oli', text):
        return "Bengkel / Servis Kendaraan"
    # Elektronik / Servis HP / Laptop / Printer
    elif re.search(r'servis hp|cell|mac|servis laptop|network|printer|service hp|notebook|laptop|komputer', text):
        return "Elektronik / Servis Gadget"
    # Tour & Travel / Rental
    elif re.search(r'travel|tour|rental|sewa motor|trip', text):
        return "Travel / Rental Kendaraan"
    # Toko Aksesoris / Pakaian / Hijab
    elif re.search(r'pakaian|fashion|baju|outfit|hijab|sprei|jilbab|aksesoris', text):
        return "Toko Pakaian / Aksesoris"
    # Toko Kosmetik / Skincare / Parfum
    elif re.search(r'kosmetik|skincare|parfum|perfume', text):
        return "Kosmetik / Parfum"
    # Apotek / Obat Herbal
    elif re.search(r'obat|apotek|herbal', text):
        return "Apotek / Obat-obatan"
    # Petshop
    elif re.search(r'petshop|hewan|pakan kucing|makanan burung', text):
        return "Petshop"
    # Studio Foto
    elif re.search(r'foto studio|foto|video|musik|studio photo|studio|fotografi', text):
        return "Studio Foto"
    # Vape / Rokok Elektrik
    elif re.search(r'vape|liquid|rokok elektrik', text):
        return "Toko Vape"
    # Koperasi / Keuangan
    elif re.search(r'koperasi|simpan pinjam|bank|bri link', text):
        return "Koperasi / Layanan Keuangan"
    # Galon / Air Minum
    elif re.search(r'air minum|galon|isi ulang', text):
        return "Isi Ulang Air Minum"
    # Kunci / Ahli Kunci / Duplikat
    elif re.search(r'kunci|duplikat kunci|ahli kunci', text):
        return "Jasa Ahli Kunci"
    # Mebel / Furnitur
    elif re.search(r'mebel|kerajinan|furnitur', text):
        return "Toko Mebel/Kerajinan"
    elif re.search(r'optik|dental|ortho', text):
        return "Kesehatan"
    elif re.search(r'adventure', text):
        return "Sewa/Perlengkapan Camp"
    # Lainnya
    else:
        return "Lainnya"

# Terapkan fungsi ke DataFrame
df['Jenis Usaha'] = df['Gabungan'].apply(klasifikasi_jenis_usaha)


# Simpan hasil ke file baru
df.to_excel("Data_Usaha_Tlogomas.xlsx", index=False)
print(df[["Nama Usaha", "Jenis Usaha"]].head(300))
```

```python
# Baca data dari Excel
df = pd.read_excel("Data_Usaha_Tlogomas.xlsx")

# Klasifikasi lanjutan untuk kost putra / putri / umum
def klasifikasi_kost_spesifik(row):
    if row['Jenis Usaha'] in ['Kost', 'Kost / Kontrakan']:
        nama = str(row['Nama Usaha']).lower()
        if 'putri' in nama or 'muslimah' in nama:
            return ['Kost Putri']
        elif 'putra' in nama:
            return ['Kost Putra']
        else:
            return ['Umum']  # Jika tidak ada kata 'putra' atau 'putri'
    return [row['Jenis Usaha']]  # Bukan kost

# Terapkan klasifikasi dan ekspansi baris jika perlu
rows = []
for _, row in df.iterrows():
    jenis_list = klasifikasi_kost_spesifik(row)
    for jenis in jenis_list:
        new_row = row.copy()
        new_row['Jenis Usaha'] = jenis
        rows.append(new_row)

# Buat DataFrame baru dari hasil ekspansi
df_expanded = pd.DataFrame(rows)

# Simpan hanya jika ada Kost Putra, Kost Putri, atau Umum
df_kost_spesifik = df_expanded[df_expanded['Jenis Usaha'].isin(['Kost Putra', 'Kost Putri', 'Umum'])]
if not df_kost_spesifik.empty:
    df_kost_spesifik.to_excel("Data_Kost_Spesifik.xlsx", index=False)
```
